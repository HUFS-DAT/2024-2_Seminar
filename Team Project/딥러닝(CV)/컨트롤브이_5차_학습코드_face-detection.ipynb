```python
import numpy as np
import cv2 as cv
from google.colab.patches import cv2_imshow
```

```python
face_cascade = cv.CascadeClassifier('haarcascade_frontalface_alt2.xml')
eye_cascade = cv.CascadeClassifier('haarcascade_eye.xml')
```

```python
img_grb = cv.imread('/content/rose_bruno.png')
img = cv.cvtColor(img_grb, cv.COLOR_BGR2RGB)
gray = cv.cvtColor(img_grb, cv.COLOR_BGR2GRAY)
```

```python
gray
```

```python
#얼굴 위치 detect
faces = face_cascade.detectMultiScale(gray, 1.3, 5)
```

```python
faces
```

```python
#앞서 detect한 얼굴 위치에서 눈 위치 detect
for (x,y,w,h) in faces:
    cv.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
    roi_gray = gray[y:y+h, x:x+w]
    roi_color = img[y:y+h, x:x+w]
    eyes = eye_cascade.detectMultiScale(roi_gray)
    for (ex,ey,ew,eh) in eyes:
        cv.rectangle(roi_color,(ex,ey),(ex+ew,ey+eh),(0,255,0),2)
```
