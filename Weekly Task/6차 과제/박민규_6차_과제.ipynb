## 가구소득정도와 스포츠관람종목1순위 카이제곱검정 실시

```python
import pandas as pd
from scipy.stats import chi2_contingency

# 교차표 생성
income_sport_crosstab = pd.crosstab(df['가구소득정도명'], df['스포츠관람종목1순위값'])
```

```python
# 카이제곱 검정 수행
chi2, p, dof, expected = chi2_contingency(income_sport_crosstab)
```

```python
# 결과 출력
print("교차표:")
pd.DataFrame(income_sport_crosstab)
```

```python
print("\n카이제곱 통계량 (Chi2):", chi2)
print("p-값:", p)
print("자유도 (Degrees of Freedom):", dof)
print("\n기대빈도 (Expected Frequencies):")
pd.DataFrame(expected, index=income_sport_crosstab.index, columns=income_sport_crosstab.columns)
```

###카이 제곱 검정 결과 p 값이 0.005로 유의수준 0.05보다 작으므로 스포츠 선호 종목과 소득 정도는 통계적으로 유의미한 관계가 있다.

## 선호종목과 관람방식은 관계가 있는가?

```python
import pandas as pd
from scipy.stats import chi2_contingency

# 교차표 생성
watch_sport_crosstab = pd.crosstab(df['1순위종목관람방법1순위유형'], df['스포츠관람종목1순위값'])
```

```python
# 카이제곱 검정 수행
chi2, p, dof, expected = chi2_contingency(watch_sport_crosstab)
```

```python
# 결과 출력
print("교차표:")
pd.DataFrame(watch_sport_crosstab)
```

```python
print("\n카이제곱 통계량 (Chi2):", chi2)
print("p-값:", p)
print("자유도 (Degrees of Freedom):", dof)
print("\n기대빈도 (Expected Frequencies):")
pd.DataFrame(expected, index=watch_sport_crosstab.index, columns=watch_sport_crosstab.columns)
```

###카이 제곱 검정 결과 p 값이 9.262486925990823e-15로 유의수준 0.05보다 작으므로 스포츠 선호 종목과 관람 방식은 통계적으로 유의미한 관계가 있다.
